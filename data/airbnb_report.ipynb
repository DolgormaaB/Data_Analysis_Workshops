```python
#df['price'] = df['price'].str.replace('$', '').astype(float)
df['price'] = df['price'].str.strip('$')
df.price.head(5)
```

```python
#df = df.astype({'price': 'float'})
df['price'] = df['price'].astype(float)
df.dtypes
```

```python
avg = df.price.mean()
```

```python
df['price'] = df['price'].fillna(avg)
df.isna().sum()
```
